```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
#Ler o arquivo
copa = pd.read_csv('WorldCups.csv')
```

```python
#teste de Leitura
copa.head()
```

```python
copa['Country'].value_counts()
```

```python
copa['Attendance'].value_counts()
```

```python
#Deletando colunas
copa.drop(['Year', 'Winner', 'Runners-Up', 'Third', 'Fourth', 'GoalsScored', 'QualifiedTeams', 'MatchesPlayed'], axis=1)
```

```python
publicoCopa = copa[['Country', 'Attendance']].head(20).set_index('Country')
publicoCopa.plot(kind='bar', figsize=(11,7), grid=True, color='Blue', legend=True)

plt.title('Attendance x Country')
plt.xlabel('Countries')
plt.ylabel('Attendance')
plt.show()
```
